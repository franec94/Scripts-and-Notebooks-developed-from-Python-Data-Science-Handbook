```python
from sklearn.datasets import make_blobs
```

```python
# Global starting variables.
seed = 0
random_state = 2

# Instantiate, or create, a random generator object.
rng = np.random.RandomState(seed=seed)
```

### Create or Derive some input data points

```python
# make_blobs?
```

```python
X, y = make_blobs(
    100, 2,
    centers=2, random_state=random_state,
    cluster_std=1.5
)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
```

### Deriving, Fitting to the Data a Gaussian naive Bayes Model

```python
model = GaussianNB()
model.fit(X, y)
```

### Predict Target Values for New, Unknown and Never Seen before Data Samples

```python
Xnew = [-6, 14] + [14, 8] * rng.rand(2000, 2)
ynew = model.predict(Xnew)
```

```python
# plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
# lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
# plt.axis(lim)
```
