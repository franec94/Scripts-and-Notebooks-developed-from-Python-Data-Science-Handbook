#### Import Modules, Packages and Third Party Libraries

```python
# Magic statements.
%matplotlib inline

# Import graph libraries.
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

# Import main modules, packages, and third party libraries.
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()

# Import scikit-learn classes: datasets.
from sklearn.datasets import load_iris

# Import scikit-learn classes: preprocessing step utility functions.
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA # Unsupervised Machine Learning tasks: feature reduction, dimensionality reduction
from sklearn.mixture import GaussianMixture # Unsupervised Machine Learning tasks: clustering
from sklearn.manifold import Isomap # Unsupervised Machine Learning tasks: feature reduction, dimensionality reduction

# Import scikit-learn classes: models
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier

# Import scikit-learn classes: Hyperparameters Validation utility functions.
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import LeavePOut
from sklearn.model_selection import LeaveOneOut

# Import scikit-learn classes: model's evaluation step utility functions.
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix
```

```python
seed = 42
random_state = 1

# Instantiate, or create, a random generator object.
rng = np.random.RandomState(seed=seed)
```

#### Data Exploration & Investigation

```python
# Load dataset 'iris'.
iris = sns.load_dataset('iris')

# Show dataset's size, that is dataset's shape in the context of dataframe.
print('Iris dataset size: ', iris.shape, 'samples x features.')
```

```python
# Show first 5 default lines, that are nothing but 5 samples within dataset 'iris'.
iris.head()
```

```python
iris.describe(include='all')
```

```python
sns.pairplot(iris, hue='species', size=1.5)
```

```python
# Extract 'feature matrix', dropping the target column.
X_iris = iris.drop('species', axis=1)

# Show X feature matrix's type.
print('X feature matrix\'s type:', type(X_iris))

# Show feature matrix's size, that is feature matrix's shape in the context of dataframe.
print('Iris Dataset, feature matrix\'s size: ', X_iris.shape, 'samples x features.')
```

```python
y_iris = iris['species']

# Show y label array's type.
print('y label array\'s type:', type(y_iris))

# Show y label array's size, that is Series's shape in the context of pandas library.
print('Iris Dataset, feature matrix\'s size: ', y_iris.shape, 'samples x features.')
```

#### Gaussian naive Bayes

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=random_state)
```

```python
model = GaussianNB()
model.fit(Xtrain, ytrain)
```

```python
y_model = model.predict(Xtest)
print('Gaussian naive Bayes accuracy score:', accuracy_score(ytest, y_model))
print(f"Gaussian naive Bayes accuracy score (percentage): {accuracy_score(ytest, y_model)*100:.2f}%")
```

#### Principal Component Analysis (PCA) for Unsupervied Machine Learning Tasks: Dimensionality reduction

```python
model = PCA(n_components=2)    # 2. Instantiate model with hyperparameters
model.fit(X_iris)              # 3. Fit to data. Notice y is not specified
X_2D = model.transform(X_iris) # 4. Transform the data to two dimensional
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]

sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False)
```

#### Gaussian Mixture Modeling (GMM) for Unsupervied Machine Learning Tasks: Clustering

```python
model = GaussianMixture(
    n_components=3,
    covariance_type='full')   # 2. Instantiate the model w/ hyperparameters.
model.fit(X_iris)             # 3. Fit the model to data. Notice y is not specified.
y_gmm = model.predict(X_iris) # 4. Determine cluster labels
```

```python
iris['cluster'] =y_gmm
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, col='cluster', fit_reg=False)
```

#### Hyperparameters and Validation - Naive Approach

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
model = KNeighborsClassifier(n_neighbors=1) # 2. Instantiate the model w/ hyperparameters.
model.fit(X, y)                             # 3. Fit the model to data. Notice y is not specified.          

y_model = model.predict(X)                  # 4. Predict sample's class labels

print('K-Neighbors Classifier accuracy score:', accuracy_score(y, y_model))
print(f"K-Neighbors Classifier accuracy score (percentage): {accuracy_score(y, y_model)*100:.2f}%")
```

```python
mat = confusion_matrix(y, y_model)

sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
```

#### Hyperparameters and Validation - Hold-Out Set Approach

```python
X1, X2, y1, y2 = train_test_split(X, y, random_state=0, train_size=0.5)
```

```python
model = KNeighborsClassifier(n_neighbors=1) # 2. Instantiate the model w/ hyperparameters.
model.fit(X1, y1)                             # 3. Fit the model to data. Notice y is not specified.          

y_model = model.predict(X2)                  # 4. Predict sample's class labels

print('K-Neighbors Classifier accuracy score:', accuracy_score(y2, y_model))
print(f"K-Neighbors Classifier accuracy score (percentage): {accuracy_score(y2, y_model)*100:.2f}%")
```

```python
mat = confusion_matrix(y2, y_model)

sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
```

#### Model Validation via Cross-Validation

```python
model = KNeighborsClassifier(n_neighbors=1) # 2. Instantiate the model w/ hyperparameters.

y2_model = model.fit(X1, y1).predict(X2)
y1_model = model.fit(X2, y2).predict(X1)

print('1.a) K-Neighbors Classifier accuracy score:', accuracy_score(y2, y2_model))
print(f"1.b) K-Neighbors Classifier accuracy score (percentage): {accuracy_score(y2, y2_model)*100:.2f}%")

print('2.a) K-Neighbors Classifier accuracy score:', accuracy_score(y1, y1_model))
print(f"2.b) K-Neighbors Classifier accuracy score (percentage): {accuracy_score(y1, y1_model)*100:.2f}%")
```

```python
model = KNeighborsClassifier(n_neighbors=1)
cross_val_score(model, X, y, cv=5)
```
