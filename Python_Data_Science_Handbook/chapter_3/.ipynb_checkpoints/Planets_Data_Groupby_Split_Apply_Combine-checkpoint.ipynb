```python
import numpy as np
import pandas as pd

import seaborn as sns
```

```python
seed = 42
rng = np.random.RandomState(seed)
```

```python
planets = sns.load_dataset('planets')
print(f"Planets data, shape: {planets.shape}")
```

```python
planets.head()
```

```python
planets.groupby('method')['orbital_period'].median()
```

```python
for (method, group) in planets.groupby('method'):
    print("{0:30s} shape={1}".format(method, group.shape))
```

```python
planets.groupby('method')['year'].describe().unstack()
```

```python
planets.groupby('method')['year'].describe().unstack().unstack()
```

```python
# Count discovered planets by method and by decade:

# print(planets['year'])

decade = 10 * (planets['year'] // 10)
# print(decade.head())
print(type(decade))

decade = decade.astype(str) + 's'
# print(decade.head())

decade.name = 'decade'
print(decade.head())

planets.groupby(['method', decade])['number'].describe().unstack().fillna(0)
# planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)
```
