```python
# Importing modules/packages:

import numpy as np
import pandas as pd

import seaborn as sns
```

```python
# Creating a random generator object:

seed = 42
rng = np.random.RandomState(seed)
```

```python
# Loading Titanic Dataset:

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby('sex')[['survived']].describe()
```

```python
titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
```

```python
titanic.groupby(['sex', 'class'])['survived'].describe()
```

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

```python
titanic.pivot_table('survived', index='sex', columns='embark_town')
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', index=['sex', age], columns='class')
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', index=['sex', age], columns='embark_town')
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', index=['sex', age], columns=[fare, 'class'])
```

```python
titanic.pivot_table('survived', index=['sex', age], columns=['embark_town', 'class'], aggfunc=['mean'])
```

```python
titanic.pivot_table(
    'survived',
    index=['embark_town'],
    columns=['class'],
    margins=True
)
```

```python
titanic.pivot_table(
    'survived',
    index=['embark_town', 'sex'],
    columns=['class'],
    margins=True
)
```
